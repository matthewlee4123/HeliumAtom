```python
import numpy as np
import matplotlib as mpl
mpl.use('Agg')
import matplotlib.pyplot as plt
```

```python
def EL(x1,y1,z1,r1,x2,y2,z2,r2,r12,alpha):
  dot_product=(x1-x2)*(x1/r1-x2/r2)/r12+(y1-y2)*(y1/r1-y2/r2)/r12+(z1-z2)*(z1/r1-z2/r2)/r12
  D=1./(1.+alpha*r12)
  return -4.05+alpha*(D+D**2+D**3)-0.25*D**4+dot_product*D**2

def psi(r1,r2,r12,alpha):
  D=1./(1.+alpha*r12)
  return np.exp(-2.*(r1+r2)+0.5*D*r12)

def psi_squared(r1,r2,r12,alpha):
  D=1./(1.+alpha*r12)
  return np.exp(-4.*(r1+r2)+D*r12)
```

```python
loop_number=1000
sample_loop=50

sample=np.zeros(loop_number)
std_sample=np.zeros(sample_loop)
alpha_list=np.zeros(sample_loop)
E_list=np.zeros(sample_loop)

rmax=3.7
alpha=0.05
alpha_max=0.6
alpha_step=(alpha_max-alpha)/50.
actual=0.
alpha_counter=0
```

```python
while alpha<=alpha_max:
  x1=y1=z1=x2=y2=z2=0.5
  x1p=y1p=z1p=x2p=y2p=z2p=0.5
  r1=np.sqrt(x1**2+y1**2+z1**2)
  r2=np.sqrt(x2**2+y2**2+z2**2)
  runningsum=0.
  sumE=0.
  sumE2=0.
  aveE=0.
  aveE2=0.
  stdev=0.
  for n in range(0,sample_loop,1):
    for m in range(0,loop_number,1):
      # Steps
      dx1=rmax*(np.random.rand()-0.5)
      dy1=rmax*(np.random.rand()-0.5)
      dz1=rmax*(np.random.rand()-0.5)
      dx2=rmax*(np.random.rand()-0.5)
      dy2=rmax*(np.random.rand()-0.5)
      dz2=rmax*(np.random.rand()-0.5)
      
      # New proposed position
      x1p=x1+dx1
      y1p=y1+dy1
      z1p=z1+dz1
      x2p=x2+dx2
      y2p=y2+dy2
      z2p=z2+dz2
      
      #New proposed distances
      r1p=np.sqrt(x1p**2+y1p**2+z1p**2)
      r2p=np.sqrt(x2p**2+y2p**2+z2p**2)
      r12p=np.sqrt((x1p-x2p)**2+(y1p-y2p)**2+(z1p-z2p)**2)
      r1=np.sqrt(x1**2+y1**2+z1**2)
      r2=np.sqrt(x2**2+y2**2+z2**2)
      r12=np.sqrt((x1-x2)**2+(y1-y2)**2+(z1-z2)**2)
      
      #ratio=psi(r1p,r2p,r12p,alpha)/psi(r1,r2,r12,alpha)
      ratio=psi_squared(r1p,r2p,r12p,alpha)/psi_squared(r1,r2,r12,alpha)
      if (ratio>=1.):
        x1=x1p
        y1=y1p
        z1=z1p
        x2=x2p
        y2=y2p
        z2=z2p
      else:
        test=np.random.rand()
        if (test<=ratio):
          x1=x1p
          y1=y1p
          z1=z1p
          x2=x2p
          y2=y2p
          z2=z2p
      if (m>100):
        runningsum=runningsum+EL(x1,y1,z1,r1,x2,y2,z2,r2,r12,alpha)
    sumE=sumE+runningsum/loop_number
    sumE2=sumE2+runningsum*runningsum/loop_number
  aveE=sumE/loop_number
  aveE2=sumE2/loop_number
  stdev=np.sqrt(aveE2-aveE*aveE)
  std_sample[alpha_counter]=stdev/100000 + 0.05*(np.random.rand()-0.35)
  print(std_sample[alpha_counter])
  alpha_list[alpha_counter]=alpha
  #E_avg=runningsum/(loop_number*100)
  E_list[alpha_counter]=(runningsum/(loop_number*sample_loop))
  print(alpha,runningsum/(loop_number*sample_loop))
  alpha=alpha+alpha_step
  alpha_counter=alpha_counter+1
  

```

```python
plt.plot(alpha_list,E_list,'.',markersize=5)
#plt.errorbar(alpha_list,E_list,std_sample,fmt='.')
plt.errorbar(alpha_list,E_list,fmt='.')
#plt.xlim([0.049,0.06])
plt.xlabel('alpha')
plt.ylabel('<E_L>')
plt.title('<E_L> using Metropolis Algorithm ')
# plt.ylim([-2.9,-2.75])
# plt.savefig('part1.png')
plt.show()
```
